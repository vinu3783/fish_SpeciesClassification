```python
# COMBINED PREPROCESSING + TRAINING PIPELINE - SINGLE CELL
# Solves the variable transfer issue by doing everything in one cell

import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import cv2
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers, callbacks
from tensorflow.keras.applications import VGG16, ResNet50, MobileNet, InceptionV3, EfficientNetB0
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.metrics import TopKCategoricalAccuracy
from sklearn.utils.class_weight import compute_class_weight
import json
import time
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

print("ğŸš€ COMBINED PREPROCESSING + TRAINING PIPELINE")
print("=" * 70)
print(f"TensorFlow: {tf.__version__} | GPU: {len(tf.config.list_physical_devices('GPU')) > 0}")

# ====================================================================
# PART 1: PREPROCESSING
# ====================================================================

print("\nğŸ“‹ PART 1: DATA PREPROCESSING")
print("=" * 50)

# Configuration
BASE_DIR = os.path.abspath('..')
DATA_DIR = os.path.join(BASE_DIR, 'data')
RESULTS_DIR = os.path.join(BASE_DIR, 'results')
PLOTS_DIR = os.path.join(RESULTS_DIR, 'plots')
MODELS_DIR = os.path.join(RESULTS_DIR, 'models')

# Create directories
for directory in [RESULTS_DIR, PLOTS_DIR, MODELS_DIR]:
    os.makedirs(directory, exist_ok=True)

# Image parameters
IMG_HEIGHT = 224
IMG_WIDTH = 224
IMG_CHANNELS = 3
BATCH_SIZE = 32
RESCALE = 1.0/255.0

# Dataset paths
TRAIN_DIR = os.path.join(DATA_DIR, 'train')
TEST_DIR = os.path.join(DATA_DIR, 'test')
VAL_DIR = os.path.join(DATA_DIR, 'val')

print(f"âš™ï¸  Configuration:")
print(f"   â€¢ Image size: {IMG_WIDTH}Ã—{IMG_HEIGHT}Ã—{IMG_CHANNELS}")
print(f"   â€¢ Batch size: {BATCH_SIZE}")
print(f"   â€¢ Data directory: {DATA_DIR}")

# Verify dataset
dataset_status = {}
for name, path in [("Train", TRAIN_DIR), ("Test", TEST_DIR), ("Val", VAL_DIR)]:
    exists = os.path.exists(path)
    dataset_status[name.lower()] = exists
    print(f"   {'âœ…' if exists else 'âš ï¸ '} {name}: {exists}")

if not dataset_status['train']:
    raise FileNotFoundError(f"Training directory required: {TRAIN_DIR}")

# Quick dataset analysis
print(f"\nğŸ“Š Dataset Analysis:")
classes = [d for d in os.listdir(TRAIN_DIR) if os.path.isdir(os.path.join(TRAIN_DIR, d))]
class_counts = {}
total_images = 0

for class_name in classes:
    class_path = os.path.join(TRAIN_DIR, class_name)
    images = [f for f in os.listdir(class_path) 
             if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))]
    class_counts[class_name] = len(images)
    total_images += len(images)

imbalance_ratio = max(class_counts.values()) / min(class_counts.values()) if class_counts else 1.0

print(f"   â€¢ Total images: {total_images:,}")
print(f"   â€¢ Classes: {len(classes)}")
print(f"   â€¢ Imbalance ratio: {imbalance_ratio:.1f}:1")

# Smart augmentation selection
if total_images < 2000:
    aug_level = "HEAVY"
elif total_images < 5000:
    aug_level = "MODERATE"
else:
    aug_level = "LIGHT"

if imbalance_ratio > 3.0:
    aug_level = "HEAVY"

print(f"ğŸ¨ Selected augmentation: {aug_level}")

# Augmentation parameters
augmentation_configs = {
    'LIGHT': {
        'rescale': RESCALE,
        'horizontal_flip': True,
        'rotation_range': 10,
        'zoom_range': 0.1
    },
    'MODERATE': {
        'rescale': RESCALE,
        'rotation_range': 20,
        'width_shift_range': 0.15,
        'height_shift_range': 0.15,
        'zoom_range': 0.15,
        'horizontal_flip': True,
        'brightness_range': [0.9, 1.1],
        'fill_mode': 'nearest'
    },
    'HEAVY': {
        'rescale': RESCALE,
        'rotation_range': 25,
        'width_shift_range': 0.2,
        'height_shift_range': 0.2,
        'shear_range': 0.15,
        'zoom_range': 0.2,
        'horizontal_flip': True,
        'brightness_range': [0.8, 1.2],
        'fill_mode': 'nearest'
    }
}

selected_aug_params = augmentation_configs[aug_level]

# Create data generators
print(f"\nğŸ­ Creating data generators...")
train_datagen = ImageDataGenerator(**selected_aug_params)
val_test_datagen = ImageDataGenerator(rescale=RESCALE)

try:
    # Training generator
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=True,
        seed=42
    )
    
    # Validation generator
    if dataset_status['val']:
        val_generator = val_test_datagen.flow_from_directory(
            VAL_DIR,
            target_size=(IMG_HEIGHT, IMG_WIDTH),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            shuffle=False
        )
    else:
        # Create validation split
        train_datagen_split = ImageDataGenerator(validation_split=0.2, **selected_aug_params)
        val_datagen_split = ImageDataGenerator(validation_split=0.2, rescale=RESCALE)
        
        train_generator = train_datagen_split.flow_from_directory(
            TRAIN_DIR,
            target_size=(IMG_HEIGHT, IMG_WIDTH),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            subset='training',
            shuffle=True,
            seed=42
        )
        
        val_generator = val_datagen_split.flow_from_directory(
            TRAIN_DIR,
            target_size=(IMG_HEIGHT, IMG_WIDTH),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            subset='validation',
            shuffle=False,
            seed=42
        )
    
    # Test generator
    if dataset_status['test']:
        test_generator = val_test_datagen.flow_from_directory(
            TEST_DIR,
            target_size=(IMG_HEIGHT, IMG_WIDTH),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            shuffle=False
        )
    else:
        test_generator = val_generator
    
    # Extract class information
    class_names = list(train_generator.class_indices.keys())
    num_classes = train_generator.num_classes
    class_indices = train_generator.class_indices
    
    print(f"âœ… Generators created:")
    print(f"   â€¢ Training: {train_generator.samples:,} images")
    print(f"   â€¢ Validation: {val_generator.samples:,} images")
    print(f"   â€¢ Test: {test_generator.samples:,} images")
    print(f"   â€¢ Classes: {num_classes}")
    
    generators_created = True
    
except Exception as e:
    print(f"âŒ Error creating generators: {e}")
    generators_created = False
    raise

# Calculate class weights
class_weights = None
use_class_weights = False

if generators_created and imbalance_ratio > 2.0:
    class_weights_array = compute_class_weight(
        'balanced',
        classes=np.unique(train_generator.classes),
        y=train_generator.classes
    )
    class_weights = dict(enumerate(class_weights_array))
    use_class_weights = True
    print(f"âš–ï¸  Class weights calculated (imbalanced dataset)")
else:
    print(f"âœ… Dataset balanced - class weights not needed")

# Quick validation
try:
    batch_x, batch_y = next(train_generator)
    train_generator.reset()
    
    validation_checks = [
        batch_x.shape[1:] == (IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS),
        0 <= batch_x.min() <= batch_x.max() <= 1.1,
        batch_y.shape[1] == num_classes
    ]
    
    if all(validation_checks):
        print("âœ… Preprocessing validation passed")
        preprocessing_valid = True
    else:
        print("âš ï¸  Some preprocessing validations failed")
        preprocessing_valid = False
        
except Exception as e:
    print(f"âŒ Preprocessing validation failed: {e}")
    preprocessing_valid = False

print(f"âœ… Preprocessing complete!")

# ====================================================================
# PART 2: MODEL TRAINING
# ====================================================================

print(f"\nğŸ“‹ PART 2: MODEL TRAINING")
print("=" * 50)

if not preprocessing_valid:
    print("âš ï¸  Preprocessing issues detected - proceeding with caution")

# Training configuration
EPOCHS = 12  # Reduced for speed
INITIAL_LR = 0.001
PATIENCE = 3

steps_per_epoch = min(train_generator.samples // BATCH_SIZE, 80)
validation_steps = min(val_generator.samples // BATCH_SIZE, 25)

print(f"âš¡ Training config:")
print(f"   â€¢ Epochs: {EPOCHS}")
print(f"   â€¢ Learning rate: {INITIAL_LR}")
print(f"   â€¢ Steps per epoch: {steps_per_epoch}")

# Model creation functions
def create_cnn_scratch(num_classes):
    """Efficient CNN from scratch"""
    return models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        layers.GlobalAveragePooling2D(),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])

def create_transfer_model(base_name, num_classes):
    """Create transfer learning model"""
    base_models = {
        'VGG16': (VGG16, (224, 224)),
        'ResNet50': (ResNet50, (224, 224)),
        'MobileNet': (MobileNet, (224, 224)),
        'InceptionV3': (InceptionV3, (299, 299)),
        'EfficientNetB0': (EfficientNetB0, (224, 224))
    }
    
    ModelClass, input_size = base_models[base_name]
    
    base_model = ModelClass(
        weights='imagenet',
        include_top=False,
        input_shape=(*input_size, 3)
    )
    base_model.trainable = False
    
    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),
        layers.Dense(num_classes, activation='softmax')
    ], name=base_name)
    
    return model, input_size

# Create all models
models_to_train = {}
model_configs = {}

print(f"\nğŸ—ï¸  Creating models...")

# CNN from scratch
models_to_train['CNN_Scratch'] = create_cnn_scratch(num_classes)
model_configs['CNN_Scratch'] = {'input_size': (224, 224), 'type': 'scratch'}

# Transfer learning models
transfer_models = ['VGG16', 'ResNet50', 'MobileNet', 'EfficientNetB0']  # Skip InceptionV3 for speed

for model_name in transfer_models:
    try:
        model, input_size = create_transfer_model(model_name, num_classes)
        models_to_train[model_name] = model
        model_configs[model_name] = {'input_size': input_size, 'type': 'transfer'}
    except Exception as e:
        print(f"âŒ {model_name} creation failed: {e}")

print(f"âœ… Created {len(models_to_train)} models")

# Training utilities
def get_callbacks(model_name):
    return [
        callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=PATIENCE,
            restore_best_weights=True,
            verbose=1
        ),
        callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.3,
            patience=2,
            min_lr=1e-7,
            verbose=1
        ),
        callbacks.ModelCheckpoint(
            f"{MODELS_DIR}/{model_name}_best.h5",
            monitor='val_accuracy',
            save_best_only=True,
            verbose=0
        )
    ]

# Train all models
print(f"\nğŸš€ Training all models...")
training_results = {}
training_histories = {}
total_start_time = time.time()

for i, (model_name, model) in enumerate(models_to_train.items(), 1):
    print(f"\nğŸ”¥ [{i}/{len(models_to_train)}] Training {model_name}")
    print(f"   Parameters: {model.count_params():,}")
    
    start_time = time.time()
    
    try:
        # Compile model
        model.compile(
            optimizer=optimizers.Adam(learning_rate=INITIAL_LR),
            loss='categorical_crossentropy',
            metrics=['accuracy', TopKCategoricalAccuracy(k=3, name='top_3_accuracy')]
        )
        
        # Train model
        history = model.fit(
            train_generator,
            epochs=EPOCHS,
            validation_data=val_generator,
            steps_per_epoch=steps_per_epoch,
            validation_steps=validation_steps,
            callbacks=get_callbacks(model_name),
            class_weight=class_weights if use_class_weights else None,
            verbose=1
        )
        
        # Record results
        training_time = time.time() - start_time
        best_val_acc = max(history.history['val_accuracy'])
        best_val_top3 = max(history.history['val_top_3_accuracy'])
        final_train_acc = history.history['accuracy'][-1]
        
        training_results[model_name] = {
            'best_val_accuracy': float(best_val_acc),
            'best_val_top3_accuracy': float(best_val_top3),
            'final_train_accuracy': float(final_train_acc),
            'training_time_minutes': training_time / 60,
            'epochs_trained': len(history.history['accuracy']),
            'total_params': model.count_params(),
            'status': 'SUCCESS'
        }
        
        training_histories[model_name] = history.history
        
        print(f"   âœ… Completed in {training_time/60:.1f} min")
        print(f"   ğŸ“Š Best Val Acc: {best_val_acc:.3f} | Top-3: {best_val_top3:.3f}")
        
        # Clear memory
        tf.keras.backend.clear_session()
        
    except Exception as e:
        training_time = time.time() - start_time
        print(f"   âŒ Failed: {str(e)[:50]}...")
        
        training_results[model_name] = {
            'status': 'FAILED',
            'error': str(e),
            'training_time_minutes': training_time / 60
        }
        
        tf.keras.backend.clear_session()

total_training_time = time.time() - total_start_time

# Results analysis
successful_models = {k: v for k, v in training_results.items() if v['status'] == 'SUCCESS'}

print(f"\nğŸ“Š TRAINING RESULTS")
print("=" * 50)
print(f"ğŸ¯ Summary:")
print(f"   â€¢ Total time: {total_training_time/60:.1f} minutes")
print(f"   â€¢ Successful: {len(successful_models)}/{len(models_to_train)} models")

if successful_models:
    # Create results table
    results_data = []
    for model_name, results in successful_models.items():
        results_data.append({
            'Model': model_name,
            'Val_Accuracy': f"{results['best_val_accuracy']:.3f}",
            'Val_Top3': f"{results['best_val_top3_accuracy']:.3f}",
            'Time_min': f"{results['training_time_minutes']:.1f}",
            'Parameters': f"{results['total_params']:,}",
            'Epochs': results['epochs_trained']
        })
    
    results_df = pd.DataFrame(results_data)
    results_df = results_df.sort_values('Val_Accuracy', ascending=False)
    
    print(f"\nğŸ† MODEL PERFORMANCE RANKING:")
    print(results_df.to_string(index=False))
    
    # Best model
    best_model_name = results_df.iloc[0]['Model']
    best_accuracy = float(results_df.iloc[0]['Val_Accuracy'])
    
    print(f"\nğŸ¥‡ BEST MODEL: {best_model_name}")
    print(f"   ğŸ“ˆ Validation Accuracy: {best_accuracy:.1%}")
    print(f"   ğŸ¯ Top-3 Accuracy: {successful_models[best_model_name]['best_val_top3_accuracy']:.1%}")

# Quick visualization
if len(successful_models) >= 2:
    print(f"\nğŸ“ˆ Creating performance visualization...")
    
    plt.figure(figsize=(12, 5))
    
    # Accuracy comparison
    plt.subplot(1, 2, 1)
    models = list(successful_models.keys())
    accuracies = [successful_models[m]['best_val_accuracy'] for m in models]
    colors = plt.cm.Set3(np.linspace(0, 1, len(models)))
    
    bars = plt.bar(models, accuracies, color=colors)
    plt.title('Model Validation Accuracy', fontweight='bold')
    plt.ylabel('Accuracy')
    plt.xticks(rotation=45)
    plt.ylim(0, 1)
    plt.grid(True, alpha=0.3)
    
    # Add labels
    for bar, acc in zip(bars, accuracies):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{acc:.3f}', ha='center', va='bottom', fontweight='bold')
    
    # Training time comparison
    plt.subplot(1, 2, 2)
    times = [successful_models[m]['training_time_minutes'] for m in models]
    bars = plt.bar(models, times, color=colors)
    plt.title('Training Time Comparison', fontweight='bold')
    plt.ylabel('Time (minutes)')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    
    for bar, time_val in zip(bars, times):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(times)*0.01,
                f'{time_val:.1f}m', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig(f"{PLOTS_DIR}/combined_training_results.png", dpi=150, bbox_inches='tight')
    plt.show()

# Save results
print(f"\nğŸ’¾ Saving results...")

final_results = {
    'metadata': {
        'created_at': datetime.now().isoformat(),
        'total_training_time_minutes': total_training_time / 60
    },
    'dataset_info': {
        'num_classes': num_classes,
        'class_names': class_names,
        'total_training_samples': train_generator.samples,
        'augmentation_level': aug_level,
        'use_class_weights': use_class_weights
    },
    'training_results': training_results,
    'best_model': {
        'name': best_model_name if 'best_model_name' in locals() else None,
        'accuracy': best_accuracy if 'best_accuracy' in locals() else None
    }
}

# Save JSON results
with open(f"{RESULTS_DIR}/combined_training_results.json", 'w') as f:
    json.dump(final_results, f, indent=2, default=str)

# Save CSV comparison
if successful_models:
    results_df.to_csv(f"{RESULTS_DIR}/model_comparison.csv", index=False)

print(f"âœ… Results saved to {RESULTS_DIR}/")

# ====================================================================
# FINAL SUMMARY
# ====================================================================

print(f"\nğŸ¯ FINAL SUMMARY")
print("=" * 70)

if successful_models:
    print(f"ğŸ‰ SUCCESS!")
    print(f"   â€¢ Preprocessing: âœ… Complete")
    print(f"   â€¢ Training: âœ… {len(successful_models)} models successful")
    print(f"   â€¢ Best Model: {best_model_name} ({best_accuracy:.1%})")
    print(f"   â€¢ Total Time: {total_training_time/60:.1f} minutes")
    
    print(f"\nğŸ“ Files Generated:")
    print(f"   â€¢ Model files: {MODELS_DIR}/*.h5")
    print(f"   â€¢ Results: {RESULTS_DIR}/combined_training_results.json")
    print(f"   â€¢ Comparison: {RESULTS_DIR}/model_comparison.csv")
    print(f"   â€¢ Plots: {PLOTS_DIR}/combined_training_results.png")
    
    print(f"\nğŸš€ READY FOR DEPLOYMENT!")
    print(f"   â€¢ Best model: {MODELS_DIR}/{best_model_name}_best.h5")
    print(f"   â€¢ Expected accuracy: ~{best_accuracy:.1%}")
    
    # Store results for potential next steps
    combined_results = {
        'preprocessing_valid': preprocessing_valid,
        'successful_models': list(successful_models.keys()),
        'best_model_name': best_model_name,
        'best_model_accuracy': best_accuracy,
        'models_directory': MODELS_DIR,
        'class_names': class_names,
        'num_classes': num_classes,
        'train_generator': train_generator,
        'val_generator': val_generator,
        'test_generator': test_generator,
        'final_results': final_results
    }
    
    globals()['combined_results'] = combined_results
    print(f"\nğŸ“¦ All results stored in 'combined_results' variable")
    
else:
    print(f"âŒ TRAINING FAILED!")
    print(f"   â€¢ Check dataset structure")
    print(f"   â€¢ Verify GPU availability") 
    print(f"   â€¢ Try reducing batch size")

print(f"\nâœ… COMBINED PIPELINE COMPLETE!")
```
